```python
# Define the path to the labels directory
labels_dir = '/kaggle/working/runs/detect/predict/labels'  # Change this to the actual path of your labels folder

# Loop through each label file and display its contents
for label_file in os.listdir(labels_dir):
    if label_file.endswith('.txt'):
        file_path = os.path.join(labels_dir, label_file)
        print(f"\nContents of {label_file}:\n")
        
        with open(file_path, 'r') as file:
            contents = file.read()
            print(contents)
```
